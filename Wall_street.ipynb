# COMBINED

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.wallstreet-online.de/optionsschein/tt09bu-microsoft-corp-call-220-bis-2021-09-hsbc'
response = requests.get(url)

response

soup = BeautifulSoup(response.text)

soup.findAll('span', href = False)

ISIN = soup.find('title', href = False).text.strip()
ISIN = ISIN[-12:]
print(ISIN)

ISIN_tag = soup.find('title', href = False).text.strip()
ISIN_tag = ISIN_tag[-12:]
print(ISIN_tag)

tag_two = soup.findAll('div', href = False)[97].text.strip()
print(tag_two)

currency = soup.findAll('span', href = False)[100].text.strip()
print(currency)

df = pd.DataFrame({'ISIN':'ISIN: ' + ISIN_tag, 'VALUE':tag_two + currency}, index= {1}) 
df.to_csv('Wall_street.csv', index=False, encoding='utf-8')
```
